Creating a 3D Surface Plot:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. Generate data for the surface plot
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Z is a function of X and Y (e.g., a Gaussian function)
Z = np.sin(np.sqrt(X**2 + Y**2))

# 2. Create the figure and 3D axes
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 3. Customize the surface plot
# Plot a 3D surface using a custom colormap and shading
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Add contour projections on the X, Y plane
ax.contour3D(X, Y, Z, 50, cmap='plasma', offset=-1, extend3d=True)

# 4. Customizing the view and plot aesthetics
ax.set_title('3D Surface Plot with Custom Colormap and Shading', fontsize=16)
ax.set_xlabel('X axis', fontsize=12)
ax.set_ylabel('Y axis', fontsize=12)
ax.set_zlabel('Z axis', fontsize=12)

# 5. Add a color bar to explain the color mapping
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

# 6. Show the plot
plt.show()
```
