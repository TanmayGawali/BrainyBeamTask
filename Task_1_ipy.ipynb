```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
import ipywidgets as widgets

# Function to update plot based on slider value
def plot_sine_wave(frequency):
    x = np.linspace(0, 2 * np.pi, 500)
    y = np.sin(frequency * x)
    plt.figure(figsize=(10, 6))
    plt.plot(x, y)
    plt.title(f'Sine wave with frequency: {frequency}')
    plt.ylim(-1.5, 1.5)
    plt.show()

# Slider widget for frequency
slider = FloatSlider(min=0.1, max=10, step=0.1, value=1)

# Linking the plot function with the slider
interact(plot_sine_wave, frequency=slider)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, Button
import ipywidgets as widgets

# Function to plot the sine wave
def plot_sine_wave(frequency, amplitude):
    x = np.linspace(0, 2 * np.pi, 500)
    y = amplitude * np.sin(frequency * x)
    plt.figure(figsize=(10, 6))
    plt.plot(x, y)
    plt.title(f'Sine wave with frequency: {frequency}, amplitude: {amplitude}')
    plt.ylim(-amplitude - 1, amplitude + 1)
    plt.show()

# Function to reset amplitude slider
def reset_amplitude(b):
    amplitude_slider.value = 1

# Creating sliders and buttons
frequency_slider = FloatSlider(min=0.1, max=10, step=0.1, value=1, description="Frequency")
amplitude_slider = FloatSlider(min=0.1, max=5, step=0.1, value=1, description="Amplitude")
reset_button = Button(description="Reset Amplitude")

# Set up button action
reset_button.on_click(reset_amplitude)

# Display widgets and plot
ui = widgets.VBox([frequency_slider, amplitude_slider, reset_button])
out = widgets.interactive_output(plot_sine_wave, {'frequency': frequency_slider, 'amplitude': amplitude_slider})

display(ui, out)
```

```python
from ipywidgets import Dropdown

# Function to change plot style
def plot_custom_wave(frequency, amplitude, wave_type):
    x = np.linspace(0, 2 * np.pi, 500)
    if wave_type == 'Sine':
        y = amplitude * np.sin(frequency * x)
    elif wave_type == 'Cosine':
        y = amplitude * np.cos(frequency * x)
    plt.figure(figsize=(10, 6))
    plt.plot(x, y)
    plt.title(f'{wave_type} wave with frequency: {frequency}, amplitude: {amplitude}')
    plt.ylim(-amplitude - 1, amplitude + 1)
    plt.show()

# Dropdown widget for selecting wave type
wave_type_dropdown = Dropdown(options=['Sine', 'Cosine'], value='Sine', description='Wave Type')

# Display widgets and plot
ui = widgets.VBox([frequency_slider, amplitude_slider, wave_type_dropdown])
out = widgets.interactive_output(plot_custom_wave, {'frequency': frequency_slider, 'amplitude': amplitude_slider, 'wave_type': wave_type_dropdown})

display(ui, out)
```
